### Mesures de Tendance Centrale

```python
print("*Moyenne (Mean)*")
print(data[['Age', 'Years of Experience', 'Salary']].mean())

print("Médiane (Median)")
print(data[['Age', 'Years of Experience', 'Salary']].median())

print("\n**Mode :**")
# Le mode peut avoir plusieurs valeurs, donc il est souvent affiché par colonne.
print(data[['Age', 'Years of Experience', 'Salary']].mode())
```

### Mesures de Dispersion

```python
print("**Écart type (Standard Deviation) :**")
print(data[['Age', 'Years of Experience', 'Salary']].std())

print("\n**Variance :**")
print(data[['Age', 'Years of Experience', 'Salary']].var())

print("\n**Étendue (Range) :**")
# L'étendue est la différence entre le maximum et le minimum
print(data[['Age', 'Years of Experience', 'Salary']].max() - data[['Age', 'Years of Experience', 'Salary']].min())
```
